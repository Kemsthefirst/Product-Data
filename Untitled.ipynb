```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mkg = pd.read_csv('marketing_campaign_dataset.xlsx - marketing_campaign_dataset.csv')
```

```python
mkg.info()
```

## Checking for missing values & Inconsistencies

```python
missing_values = mkg.isnull().sum()

categorical_cols = ['Company', 'Campaign_Type', 'Target_Audience', 'Channel_Used', 'Location', 'Customer_Segment']
unique_values = {col: mkg[col].unique() for col in categorical_cols}

print("Missing Values:\n", missing_values)
print("\nUnique Values in Categorical Columns:\n", unique_values)
```

## Basic statistics (mean, median, standard deviation).

```python
mkg.describe()
```

## Removing Dollar signs and Commas and converting Acquisition Column to a float

```python
mkg['Acquisition_Cost'] = mkg['Acquisition_Cost'].replace('[\$,]', '', regex=True).astype(float)
```

## Unique target audiences and marketing channels

```python
print("Unique Target Audiences:\n", mkg['Target_Audience'].value_counts())
print("\nUnique Marketing Channels:\n", mkg['Channel_Used'].value_counts())
```

## Outliers in Numerical Columns (impressions, clicks, spend)

```python
outlier_cols = ['Impressions', 'Clicks', 'Acquisition_Cost']

plt.figure(figsize=(12, 6))
for i, col in enumerate(outlier_cols, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=mkg[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

## Comparing campaign performance across different channels.

```python
channel_performance = mkg.groupby('Channel_Used').agg({
    'Conversion_Rate': 'mean',
    'ROI': 'mean',
    'Clicks': 'sum',
    'Impressions': 'sum'
}).sort_values(by='ROI', ascending=False)

plt.figure(figsize=(10, 5))
ax = sns.barplot(x=channel_performance.index, y=channel_performance['ROI'], color='royalblue')

for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width()/2, p.get_height()), 
                ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xticks(rotation=45)
plt.ylabel('ROI')
plt.title('ROI Across Marketing Channels') 
plt.show()
```

## Calculating CTR, CPC, and conversion rates

```python
mkg['CTR'] = (mkg['Clicks'] / mkg['Impressions']) * 100
mkg['CPC'] = mkg['Acquisition_Cost'] / mkg['Clicks']

plt.figure(figsize=(8, 5))
sns.scatterplot(x=mkg['CTR'], y=mkg['CPC'], alpha=0.7, color='royalblue')
plt.xlabel('Click-Through Rate (%)')
plt.ylabel('Cost Per Click ($)')
plt.title('CTR vs CPC')
plt.savefig('CTR vs CPC.png') 
plt.show()

print("CTR, CPC & Conversion Rate Data:\n")
print(mkg[['CTR', 'CPC', 'Conversion_Rate']].head())
```

## Identifying high-performing and underperforming campaigns

```python
campaign_performance = mkg.groupby('Campaign_Type')['ROI'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.lineplot(x=campaign_performance.index, y=campaign_performance.values, marker='o', color='royalblue')

for x, y in zip(campaign_performance.index, campaign_performance.values):
    plt.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

plt.xlabel('Campaign Type')
plt.ylabel('Average ROI')
plt.title('ROI Across Campaign Types')
plt.xticks(rotation=45)
plt.show()
```

## Location based trends

```python
location_performance = mkg.groupby('Location').agg({
    'Conversion_Rate': 'mean',
    'ROI': 'mean',
    'Clicks': 'sum',
    'Impressions': 'sum'
}).sort_values(by='ROI', ascending=False)

plt.figure(figsize=(8, 4))
sns.heatmap(location_performance[['ROI']], cmap='coolwarm', annot=True, fmt='.2f', cbar=True)
plt.title('ROI Across Locations')
plt.show()
```
